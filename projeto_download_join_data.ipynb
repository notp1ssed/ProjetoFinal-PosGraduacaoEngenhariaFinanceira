# Separar DFs em periodica e cumulativa

Sim, separar os dados em dois DataFrames (um para as variáveis cumulativas e outro para as variáveis periódicas) pode ser uma abordagem eficiente e organizada. Isso facilita o tratamento específico que cada grupo de variáveis precisa antes de integrá-los novamente para a análise final de regressão.

Aqui está um workflow sugerido:

### Workflow Sugerido:

1. **Separação dos DataFrames:**
   - **DF Cumulative:** Crie um DataFrame com as 15 variáveis acumuladas.
   - **DF Periodic:** Crie um DataFrame com as 5 variáveis periódicas.

2. **Tratamento das Variáveis Cumulativas:**
   - **Transformação Percentual:** Se necessário, transforme as variáveis acumuladas em variações percentuais em relação a um período inicial ou a outro ponto de referência.
   - **Transformação Logarítmica:** Aplique transformações logarítmicas nas variáveis acumuladas, se apropriado, para linearizar relações ou reduzir assimetria.
   - **Verificação de Outliers:** Identifique e trate outliers se necessário, já que os dados transformados podem ainda conter valores extremos que afetam a regressão.
   - **Normalização:** Se for necessário, normalize as variáveis cumulativas para ter uma escala comparável às periódicas.

3. **Tratamento das Variáveis Periódicas:**
   - **Agregação Temporal (se necessário):** Se o contexto permitir, agregue os dados diários em médias, somas ou outras estatísticas que melhor se ajustem à sua análise.
   - **Verificação de Outliers:** Similar ao tratamento das variáveis cumulativas, verifique e trate outliers nas variáveis periódicas.
   - **Normalização:** Normalize as variáveis periódicas para garantir consistência na escala dos dados.

4. **Combinação dos DataFrames:**
   - **Juntar os DataFrames:** Após o tratamento, junte os DataFrames novamente em um único DataFrame.
   - **Checagem Final:** Verifique se todas as variáveis estão corretamente escaladas e que não há problemas de multicolinearidade entre elas.

5. **Análise de Regressão:**
   - **Verificação de Multicolinearidade:** Execute análises de multicolinearidade, como o cálculo do VIF, para garantir que as variáveis não estão altamente correlacionadas.
   - **Modelagem:** Escolha o modelo de regressão apropriado (linear, lasso, ridge, etc.) com base nas características dos dados.
   - **Validação:** Valide o modelo com técnicas de validação cruzada ou utilizando um conjunto de dados de teste separado.

6. **Interpretação dos Resultados:**
   - **Análise dos Coeficientes:** Interprete os coeficientes do modelo para entender a influência de cada variável.
   - **Diagnósticos:** Execute diagnósticos de modelo, como análise dos resíduos, para garantir a adequação do modelo.

### Benefícios Desta Abordagem:
- **Organização:** Separar os DataFrames permite que você trate de forma eficiente e clara as necessidades específicas de cada grupo de variáveis.
- **Flexibilidade:** Facilita o ajuste e transformação de variáveis, permitindo testes e iterações antes da combinação final.
- **Clareza na Análise:** Mantém a análise clara, permitindo identificar rapidamente problemas específicos (como outliers ou multicolinearidade) em cada grupo antes de prosseguir com a análise integrada.

Essa abordagem modular facilita o tratamento adequado dos dados e assegura que as variáveis estejam prontas para uma análise de regressão robusta.

```python
pct_log_week_cumulative_df.tail()
```

90% winsorization means the replacement of the top 5% and bottom 5% of the data. The top 5% of the data is replaced by the value of the data at the 95th percentile and the value of the bottom 5% of the data is replaced by the value of the data at the 5th percentile. 

```python
winsorized_pct_log_week_cumulative_df = pct_log_week_cumulative_df.apply(lambda col: winsorize(col, (0.05, 0.05)))
winsorized_pct_log_week_cumulative_df
```
